## Таблица post_text_df

1. Получаем CLS-эмбеддинги
2. Уменьшаем размерность с помощью PCA
3. Кластеризуем через KMeans
4. Новые фичи: номер кластера (TextCluster), расстояние до центра кластера (DistanceToCentre), РСА компоненты (bert_pca) 

```python
import torch
from torch.utils.data import Dataset, DataLoader
from tqdm.auto import tqdm
from transformers import AutoTokenizer, BertModel, RobertaModel, DistilBertModel
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
```

```python
def get_model(model_name: str):
    assert model_name in ['bert', 'roberta', 'distilbert']

    checkpoint_names = {
        'bert': 'bert-base-cased',
        'roberta': 'roberta-base',
        'distilbert': 'distilbert-base-cased'
    }
    model_classes = {
        'bert': BertModel,
        'roberta': RobertaModel,
        'distilbert': DistilBertModel
    }

    tok = AutoTokenizer.from_pretrained(checkpoint_names[model_name])
    mdl = model_classes[model_name].from_pretrained(checkpoint_names[model_name])
    return tok, mdl

### функция реализована так, потому что взял ее с прошлого задания на курсе, где нужно было делать
### повторяющиеся действия для каждой из 3 моделей

MODEL_NAME = 'roberta'
tokenizer, bert_model = get_model(MODEL_NAME)
```

```python
### максимальная длина поста
query_5 = """SELECT MAX(
         CASE 
           WHEN COALESCE(text, '') = '' THEN 0
           ELSE array_length(
                  regexp_split_to_array(trim(regexp_replace(text, '\s+', ' ', 'g')), ' '),
                  1
                )
         END
       ) AS max_words
FROM public.post_text_df"""

max_len = pd.read_sql(query_5, con=engine)
print(max_len)
```

```python
lengths = []
for t in post_text_df["text"].fillna("").astype(str):
    ids = tokenizer(t, truncation=False)["input_ids"]
    lengths.append(len(ids))

import numpy as np
print("max tokens:", max(lengths))
for q in [0.5, 0.9, 0.95, 0.99]:
    print(q, int(np.quantile(lengths, q)))
```

```python
class PostTextDataset(Dataset):
    def __init__(self, texts):
        self.texts = texts

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return self.texts[idx]


def collate_texts(texts, tokenizer, max_length=512):
    return tokenizer(
        list(texts),
        padding=True,
        truncation=True,
        max_length=max_length,
        return_tensors="pt",
        return_token_type_ids=False
    )


@torch.inference_mode()
def get_cls_embeddings(model, loader, tokenizer, device, max_length=256):
    model.eval()
    out = []

    for batch_texts in tqdm(loader, desc="BERT embeddings"):
        batch = collate_texts(batch_texts, tokenizer, max_length=max_length)
        batch = {k: v.to(device) for k, v in batch.items()}

        emb = model(**batch).last_hidden_state[:, 0, :]
        out.append(emb.cpu())

    return torch.cat(out, dim=0)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("device:", device)

bert_model = bert_model.to(device)

post_texts = post_text_df["text"].fillna("").astype(str).tolist()

ds = PostTextDataset(post_texts)
loader = DataLoader(ds, batch_size=32, shuffle=False, num_workers=0)

embeddings = get_cls_embeddings(bert_model, loader, tokenizer, device, max_length=256) \
    .numpy().astype(np.float32)

print("embeddings shape:", embeddings.shape)
```

```python
pca_dim = 50
n_clusters = 15

centered = embeddings - embeddings.mean(axis=0, keepdims=True)

pca = PCA(n_components = pca_dim, random_state=0)
pca_decomp = pca.fit_transform(centered)

kmeans = KMeans(n_clusters=n_clusters, random_state=0, n_init=10).fit(pca_decomp)

bert_features = pd.DataFrame({"post_id": post_text_df["post_id"].values})
bert_features["TextCluster"] = kmeans.labels_.astype(int)

dists = kmeans.transform(pca_decomp)
dists_columns = [f"DistanceToCluster_{i}" for i in range(n_clusters)]
bert_features[dists_columns] = dists

pca_columns = [f"bert_pca_{i}" for i in range(pca_dim)]
bert_features[pca_columns] = pca_decomp

bert_features.head()
```

```python
df = post_text_df.merge(bert_features, on="post_id", how="left", validate="one_to_one")

df["TextCluster"] = df["TextCluster"].fillna(-1).astype(int)
cluster_ohe = pd.get_dummies(df["TextCluster"], prefix="text_cluster", dtype=int)

topic_ohe = pd.get_dummies(df["topic"], prefix="topic", dtype=int)

post_features = pd.concat(
    [
        df.drop(columns=["TextCluster", "topic"], errors="ignore"),  
        cluster_ohe,
        topic_ohe,
    ],
    axis=1
)

post_features.head()
```

```python
post_features["text_len"] = post_features["text"].apply(lambda x: len(str(x).split()) if pd.notna(x) else 0)
```

```python
post_features["unique_word_count"] = post_features["text"].apply(lambda x: len(set(str(x).split())) if pd.notna(x) else 0)
```

```python
post_features["text_diversity"] = post_features.apply(lambda row: row["unique_word_count"]/row["text_len"]
                                                             if row["text_len"] > 0 else 0, axis=1)
post_features
```

```python
post_features = post_features.drop(columns=["text"])
```

## Таблица feed_data

```python
with engine.connect() as conn:
    conn.exec_driver_sql("ROLLBACK")

N = 1000000

q = f"""
SELECT "timestamp", user_id, post_id, target
FROM public.feed_data
WHERE action = 'view'
ORDER BY "timestamp" DESC
LIMIT {N}
"""

feed_data = pd.read_sql(q, con=engine, parse_dates=['timestamp'])
```

```python
feed_data
```

```python
feed_data["hour"] = feed_data["timestamp"].dt.hour
feed_data["weekday"] = feed_data["timestamp"].dt.weekday
feed_data["is_weekend"] = (feed_data["weekday"] >= 5).astype(int)

cutoff = feed_data["timestamp"].quantile(0.8)
train = feed_data[feed_data["timestamp"] < cutoff].copy()
test = feed_data[feed_data["timestamp"] >= cutoff].copy()

train_users = set(train["user_id"])
test = test[test["user_id"].isin(train_users)].copy()

```

```python
post_stats = (train.groupby("post_id")
    .agg(post_views = ("target", "count"), 
         post_likes = ("target", "sum"))
            .reset_index()
             )

post_stats["post_ctr"] = post_stats["post_likes"] / post_stats["post_views"]
post_stats = post_stats.fillna(0)

user_stats = (train.groupby("user_id")
             .agg(user_views = ("target", "count"),
                  user_likes = ("target", "sum"))
             .reset_index())

user_stats["user_ctr"] = user_stats["user_likes"] / user_stats["user_views"]
user_stats = user_stats.fillna(0)

post_features = post_features.merge(post_stats, on="post_id", how="left")
user_features = user_features.merge(user_stats, on="user_id", how="left")

post_features[["post_views", "post_likes", "post_ctr"]] = post_features[["post_views", "post_likes", "post_ctr"]].fillna(0)
user_features[["user_views", "user_likes", "user_ctr"]] = user_features[["user_views", "user_likes", "user_ctr"]].fillna(0)
```

# Мерджим таблицы

```python
train_full = (train
              .merge(user_features, on="user_id", how="left")
              .merge(post_features, on="post_id", how="left"))

test_full = (test
            .merge(user_features, on="user_id", how="left")
            .merge(post_features, on="post_id", how="left"))
```

```python
y_train = train_full["target"].astype(int)
y_test = test_full["target"].astype(int)
```

```python
drop_cols = ["timestamp","user_id","post_id","target", "unique_word_count"]
X_train = train_full.drop(columns=[c for c in drop_cols if c in train_full.columns]).fillna(0)
X_test   = test_full.drop(columns=[c for c in drop_cols if c in test_full.columns]).fillna(0)
```

```python
print(list(X_train.columns))
```

```python
from catboost import CatBoostClassifier

pos_rate = float(y_train.mean())
pos_w = (1 - pos_rate) / max(pos_rate, 1e-8)

model = CatBoostClassifier(
    iterations=3000,
    learning_rate=0.03,
    depth=6,
    loss_function='Logloss',
    eval_metric='AUC',
    class_weights=[1.0, pos_w],
    l2_leaf_reg=5,          
    random_seed=1,
    early_stopping_rounds=300,
    verbose=200
)


model.fit(X_train, y_train,
          eval_set=(X_test,  y_test))
```

```python
y_pred = model.predict_proba(X_test)[:, 1]
y_pred
```

```python
eval_df = pd.DataFrame({
    "user_id": test_full["user_id"].values,
    "target": y_test.values,
    "pred": y_pred
})

top5 = (eval_df.sort_values(['user_id', 'pred'], ascending=[True, False])
                .groupby('user_id', as_index=False)
                .head(5))

hitrate5 = top5.groupby('user_id')['target'].max().mean()
print("HitRate@5:", round(hitrate5, 4))
```
