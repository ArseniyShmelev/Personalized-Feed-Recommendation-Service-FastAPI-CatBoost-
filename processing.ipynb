### Мерджим таблицы ###

```python
train_full = (train
              .merge(user_features, on="user_id", how="left")
              .merge(post_features, on="post_id", how="left"))

test_full = (test
            .merge(user_features, on="user_id", how="left")
            .merge(post_features, on="post_id", how="left"))
```

```python
train_full
```

```python
test_full
```

```python
y_train = train_full["target"].astype(int)
y_test = test_full["target"].astype(int)
```

```python
drop_cols = ["timestamp","user_id","post_id","target", "unique_word_count"]
X_train = train_full.drop(columns=[c for c in drop_cols if c in train_full.columns]).fillna(0)
X_test   = test_full.drop(columns=[c for c in drop_cols if c in test_full.columns]).fillna(0)
```

```python
X_train.columns
```

```python
from catboost import CatBoostClassifier

pos_rate = float(y_train.mean())
pos_w = (1 - pos_rate) / max(pos_rate, 1e-8)

model = CatBoostClassifier(
    iterations=3000,
    learning_rate=0.03,
    depth=6,
    loss_function='Logloss',
    eval_metric='AUC',
    class_weights=[1.0, pos_w],
    l2_leaf_reg=5,          
    random_seed=1,
    early_stopping_rounds=300,
    verbose=200
)


model.fit(X_train, y_train,
          eval_set=(X_test,  y_test))
```

```python
y_pred = model.predict_proba(X_test)[:, 1]
y_pred
```

```python
eval_df = pd.DataFrame({
    "user_id": test_full["user_id"].values,
    "target": y_test.values,
    "pred": y_pred
})

top5 = (eval_df.sort_values(['user_id', 'pred'], ascending=[True, False])
                .groupby('user_id', as_index=False)
                .head(5))

hitrate5 = top5.groupby('user_id')['target'].max().mean()
print("HitRate@5:", round(hitrate5, 4))
```
